```python
dataset = dataset.replace(0, np.nan)
```

```python
dataset.isna().sum()
```

```python
#Above analysis provide that SkinThickness and Insulin has got 0 values respectivly 227 and 374 so we have to 
#ignore the both columns due to 0 values

dataset = dataset.replace(np.nan, 0)
```

```python
dataset.isna().sum()
```

```python
#Select only the useful features

useful_FeatureSet = ['Pregnancies','Glucose','BloodPressure','BMI','DiabetesPedigreeFunction','Age','class']
dataset_use = dataset[useful_FeatureSet]

dataset_use.head()
```

```python
dataset_use['BMI'] = dataset_use['BMI'].replace(0,dataset_use['BMI'].median())
dataset_use['BloodPressure'] = dataset_use['BloodPressure'].replace(0,dataset_use['BloodPressure'].median())
dataset_use['Glucose'] = dataset_use['Glucose'].replace(0,dataset_use['Glucose'].median())
```

```python
arrayset = dataset_use.values
X_data = arrayset[:, 0:6]
Y_data = arrayset[:, 6]
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data, test_size=0.2, random_state=7)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
from sklearn.ensemble import RandomForestClassifier

RandomForestClassifier = RandomForestClassifier()
RandomForestClassifier.fit(X_train,Y_train)
```

```python
RandomForestClassifier_Result = RandomForestClassifier.predict(X_test)
```

```python
RandomForestClassifier_Result = RandomForestClassifier.predict(X_test)

msg = "%s: (%f)" % ("Accuracy Score Random Forest Classifier", metrics.accuracy_score
                    (Y_test, RandomForestClassifier_Result)*100)
print(msg)
```

```python
KNeighborsClassifier = KNeighborsClassifier()
KNeighborsClassifier.fit(X_train,Y_train)
```

```python
KNeighborsClassifier_Result = KNeighborsClassifier.predict(X_test)
msg = "%s: (%f)" % ("Accuracy Score K-Nearest Neighbor (KNN)", metrics.accuracy_score
                    (Y_test, KNeighborsClassifier_Result)*100)
print(msg)
```

```python
LogisticRegression_model = LogisticRegression(solver='liblinear', multi_class='ovr')
LogisticRegression_model.fit(X_train,Y_train)
```

```python
LogisticRegression_Result = LogisticRegression_model.predict(X_test)

msg = "%s: (%f)" % ("Accuracy Score Logistic Regression", metrics.accuracy_score(Y_test,
LogisticRegression_Result)*100)
print(msg)
```

```python
LinearDiscriminantAnalysis = LinearDiscriminantAnalysis()
LinearDiscriminantAnalysis.fit(X_train,Y_train)
```

```python
LinearDiscriminantAnalysis_Result = LinearDiscriminantAnalysis.predict(X_test)

msg = "%s: (%f)" % ("Accuracy Score Linear Discriminant Analysis", metrics.accuracy_score
                    (Y_test, LinearDiscriminantAnalysis_Result)*100)
print(msg)
```

```python
DecisionTreeClassifier = DecisionTreeClassifier()
DecisionTreeClassifier.fit(X_train,Y_train)
```

```python
DecisionTreeClassifier_Result = DecisionTreeClassifier.predict(X_test)

msg = "%s: (%f)" % ("Accuracy Score Decision Tree Classifier", metrics.accuracy_score
                    (Y_test, DecisionTreeClassifier_Result)*100)
print(msg)
```
